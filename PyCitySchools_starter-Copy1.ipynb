```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data= pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data,school_data,how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

## Local Government Area Summary

```python
# Calculate the Totals (Schools and Students)
unique_schools = school_data_complete_df["school_name"].unique()
school_count = len(unique_schools)
print(f"Total number of unique schools: {school_count}")

student_count = school_data_complete_df["student_name"].count()
print(f"Total student counts is: {student_count}")

# Calculate the Total Budget
total_budget = school_data["budget"].sum()
print(f"Total budget across all 15 schools : {total_budget}")
```

```python
# Calculate the Average Scores
average_maths_score = round(school_data_complete_df["maths_score"].mean(),2)
print(f"Average Match scores : {average_maths_score}")

average_reading_score = round(school_data_complete_df["reading_score"].mean(),2)
print(f"Average Reading scores : {average_reading_score}")
```

```python
# Calculate the Percentage Pass Rates
passing_maths_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)].count()["student_name"]
passing_maths_percentage = round(passing_maths_count / float(student_count) * 100,2)
print(f"passing_maths_percentage is : {passing_maths_percentage}%")

#passing_reading_percentage
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
passing_reading_percentage = round(passing_reading_count / float(student_count) * 100,2)
print(f"passing_reading_percentage is : {passing_reading_percentage}%")

#passing_maths_reading_percentage
passing_maths_reading_count =school_data_complete_df[(school_data_complete_df["maths_score"] >= 50) & \
                                                     (school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
passing_maths_reading_percentage = round(passing_maths_reading_count /float(student_count) * 100 ,2)
print(f"passing_maths_reading_percentage is : {passing_maths_reading_percentage}%")

#overall_passing_rate
overall_passing_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50) | \
                                                     (school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
overall_passing_rate = round(overall_passing_count/float(student_count)*100,2)
print(f"overall_passing_rate is: {overall_passing_rate}%")
```

```python
# Convert to DataFrame
area_summary_df = pd.DataFrame({

    "Local Government Area Summary" :["Total School","Total Students","Total Budget",
                                      "Average Maths Score","Average Reading Score",
                                      "% of Students passed Math","% of Students passed Reading",
                                      "% of Students passed Math & Reading"],
    " " : ["{:,}".format(school_count),
           "{:,}".format(student_count),
           "${:,.2f}".format(total_budget),
           "{:,.2f}".format(average_maths_score),
            "{:,.2f}".format(average_reading_score),
           "{:,.2f}%".format(passing_maths_percentage),
           "{:,.2f}%".format(passing_reading_percentage),
           "{:,.2f}%".format(passing_maths_reading_percentage)
          ]
                             })

# Formatting- formating works with above code.... ned to correct below code
#area_summary_df["Total School"] = area_summary_df["Total School"].astype(float).map("{:,}".format)
#area_summary_df["Total Students"] = area_summary_df["Total Students"].astype(float).map("{:,}".format)
#area_summary_df["Total Budget"] = area_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
#area_summary_df["% of Students passed Math"] = area_summary_df["% of Students passed Math"].astype(float).map("{:,.2f}%".format)
#area_summary_df["% of Students passed Reading"] = area_summary_df["% of Students passed Reading"].astype(float).map("{:,.2f}%".format)
#area_summary_df["% of Students passed Math & Reading"] = area_summary_df["% of Students passed Math & Reading"].astype(float).map("{:,.2f}%".format)

# Display the DataFrame
#print(area_summary_df.to_string(index=False))
area_summary_df
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_data_complete_df.set_index(["school_name"])["type"]

school_grouped = school_data_complete_df.groupby("school_name")

# Calculate the total student count per school from school_data
per_school_counts = school_grouped["student_name"].count()

# Calculate the total school budget and per capita spending per school from school_data
per_school_budget = school_grouped["budget"].mean()
per_school_capita = per_school_budget / per_school_counts

# Calculate the average test scores per school from school_data_complete
per_school_maths = school_grouped["maths_score"].mean()
per_school_reading = school_grouped["reading_score"].mean()
```

```python
# Get the students who passed maths and passed reading by creating separate filtered DataFrames from school_data_complete.
school_passing_maths = school_data_complete_df[school_data_complete_df["maths_score"] >= 50]
school_passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 50]

# Get the students who passed both reading and maths in a separate DataFrame from school_data_complete.
school_passing_both = school_data_complete_df[(school_data_complete_df["maths_score"] >=50) &\
                                              (school_data_complete_df["reading_score"] >=50)]
```

```python
#  Calculate the Percentage Pass Rates
  
per_school_passing_maths = round((school_data_complete_df["maths_score"] >= 50)
                                 .groupby(school_data_complete_df["school_name"]).mean() * 100, 2)

per_school_passing_reading = round((school_data_complete_df["reading_score"] >= 50).
                                   groupby(school_data_complete_df["school_name"]).mean() * 100, 2)

overall_passing_rate = round((per_school_passing_maths + per_school_passing_reading) / 2, 2)
```

```python
# Convert to DataFrame
school_types = school_data_complete_df.drop_duplicates(subset=["school_name"]).set_index(["school_name"])["type"]

per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Maths Score": per_school_maths,
    "Average Reading Score": per_school_reading,
    "% Passing Maths": per_school_passing_maths,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Maths Scores by Year

```python
# Create data series of scores by year levels using conditionals
year_nine = school_data_complete_df[(school_data_complete_df["year"] == 9)]
year_ten = school_data_complete_df[(school_data_complete_df["year"] == 10)]
year_eleven = school_data_complete_df[(school_data_complete_df["year"] == 11)]
year_twelve = school_data_complete_df[(school_data_complete_df["year"] == 12)]
                                             
# Group each by school name
year_nine_scores = year_nine.groupby("school_name")["maths_score"].mean()
year_ten_scores =year_ten.groupby("school_name")["maths_score"].mean()
year_eleven_scores =year_eleven.groupby("school_name")["maths_score"].mean()
year_twelve_scores =year_twelve.groupby("school_name")["maths_score"].mean()

# Combine series into single DataFrame
maths_scores_by_year = pd.DataFrame({  "Year 9":year_nine_scores,
                                        "Year 10":year_ten_scores,
                                        "Year 11":year_eleven_scores,
                                        "Year 12":year_twelve_scores
                                    })

# Minor data wrangling
maths_scores_by_year.index.name = None

# Display the DataFrame
maths_scores_by_year
```

## Reading Score by Year

```python
# Create data series of scores by year levels using conditionals
year_nine = school_data_complete_df[(school_data_complete_df["year"] == 9)]
year_ten = school_data_complete_df[(school_data_complete_df["year"] == 10)]
year_eleven = school_data_complete_df[(school_data_complete_df["year"] == 11)]
year_twelve = school_data_complete_df[(school_data_complete_df["year"] == 12)]

# Group each by school name
year_nine_scores = year_nine.groupby("school_name")["reading_score"].mean()
year_ten_scores =year_ten.groupby("school_name")["reading_score"].mean()
year_eleven_scores =year_eleven.groupby("school_name")["reading_score"].mean()
year_twelve_scores =year_twelve.groupby("school_name")["reading_score"].mean()

# Combine series into single DataFrame
reading_scores_by_year = pd.DataFrame({  "Year 9":year_nine_scores,
                                        "Year 10":year_ten_scores,
                                        "Year 11":year_eleven_scores,
                                        "Year 12":year_twelve_scores
                                    })

# Minor data wrangling
reading_scores_by_year.index.name = None

# Display the DataFrame
reading_scores_by_year
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
#  This step can be skipped but its best to make a copy.
school_spending_df = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Maths Score": per_school_maths,
    "Average Reading Score": per_school_reading,
    "% Passing Maths": per_school_passing_maths,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})
```

```python
# Categorise spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(
                                                            school_spending_df["Per Student Budget"],
                                                            bins=spending_bins,labels=group_names,
                                                            include_lowest=True
                                                            )

school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"],observed=False)["Average Maths Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"],observed=False)["Average Reading Score"].mean()
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"],observed=False)["% Passing Maths"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"],observed=False)["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"],observed=False)["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
                                    "Average Maths Score":spending_maths_scores,
                                    "Average Reading Score":spending_reading_scores,
                                    "% Passing Maths":spending_passing_maths,
                                    "% Passing Reading":spending_passing_reading,
                                    "% Overall Passing":overall_passing_spending
                                        })

# Display results
spending_summary
```
